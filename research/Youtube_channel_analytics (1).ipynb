```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# droping the columns with empty values
dataset.drop(columns=['annotationClickThroughRate','annotationCloseRate','annotationImpressions','annotationClickableImpressions','annotationClosableImpressions','annotationClicks','annotationCloses'],inplace=True)
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.describe()
```

```python
feature = dataset.drop(columns=['views','video_id'])
target = dataset['views']
```

```python
feature.head()
```

```python
feature['day'].dtype
```

```python
feature['day'] = pd.to_datetime(feature['day'])
feature['year'] = feature['day'].dt.year
feature['month'] = feature['day'].dt.month
feature['day_of_month'] = feature['day'].dt.day
```

```python
feature.head()
```

```python
feature=feature.drop(columns=['day'],axis=1)
```

```python
feature.head()
```

```python
target.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(feature,target,test_size=0.2,random_state=2)
```

```python
print(feature.shape,X_train.shape,X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

X_train = scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
X_train.shape
```

```python
print(X_train,X_test)
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR

# models={
#     'Linear Regression': LinearRegression(),
#     'Ridge Regression': Ridge(),
#     'Lasso Regression': Lasso(),
#     'Decision Tree': DecisionTreeRegressor(),
#     'Random Forest': RandomForestRegressor(),
#     'Support Vector Regression': SVR(kernel='rbf')
# }

# # Evalute each model for cross-validation.
# results={}
# for model_names, model in models.items():
#   scores= cross_val_score(model,X_train,y_train, cv=5, scoring='neg_mean_squared_error')
#   results[model_names] = (-scores.mean(), scores.std()) # Negative MSE to MSE

# # Print the results
# for model_name, (mean_mse, std_mse) in results.items():
#   print(f"{model_name}: Mean MSE = {mean_mse:.2f}, Std MSE = {std_mse:.2f}")

# Running them all together is taking way long time then i expected soo thought of skipping to running it individually.
```

```python
linear_reg = LinearRegression()
scores = cross_val_score(linear_reg, X_train, y_train, cv=5, scoring="neg_mean_squared_error")
mean_mse = -scores.mean()
std_mse = scores.std()
print(f"Linear Regression: Mean MSE: {mean_mse:.2f}, Std MSE: {std_mse:.2f}")
```

```python
lasso_reg = Lasso()
scores = cross_val_score(lasso_reg, X_train, y_train, cv=5, scoring="neg_mean_squared_error")
mean_mse = -scores.mean()
std_mse = scores.std()
print(f"Lasso Regression: Mean MSE: {mean_mse:.2f}, Std MSE: {std_mse:.2f}")
```

```python
rigde_reg = Ridge()
scores = cross_val_score(rigde_reg, X_train, y_train, cv=5, scoring="neg_mean_squared_error")
mean_mse = -scores.mean()
std_mse = scores.std()
print(f"Rigde Regression: Mean MSE: {mean_mse:.2f}, Std MSE: {std_mse:.2f}")
```

```python
dtr_reg = DecisionTreeRegressor()
scores = cross_val_score(dtr_reg, X_train, y_train, cv=5, scoring="neg_mean_squared_error")
mean_mse = -scores.mean()
std_mse = scores.std()
print(f"Decision Tree Regression: Mean MSE: {mean_mse:.2f}, Std MSE: {std_mse:.2f}")
```

```python
rfr_reg = RandomForestRegressor(n_estimators=50, max_depth=10, n_jobs=-1) # its taking a very long time soo I added the n_estimators and max_depth parameter.
scores = cross_val_score(rfr_reg, X_train, y_train, cv=5, scoring="neg_mean_squared_error")
mean_mse = -scores.mean()
std_mse = scores.std()
print(f"Random Forest Regression: Mean MSE: {mean_mse:.2f}, Std MSE: {std_mse:.2f}")
```

```python
# svr_reg = SVR(kernel='linear', C=0.1, epsilon=0.2)
# scores = cross_val_score(svr_reg, X_train, y_train, cv=5, scoring="neg_mean_squared_error")
# mean_mse = -scores.mean()
# std_mse = scores.std()
# print(f"Support Vector Regression: Mean MSE: {mean_mse:.2f}, Std MSE: {std_mse:.2f}")
```

So from the evaluation of multiple regression models we can conclude that the RandomForestRegressor performs the best out of all the models.

So I will be using the RandomForestRegressor for model traning.

```python
model = RandomForestRegressor(n_estimators=200, random_state=0, oob_score=True)
model.fit(X_train,y_train)
```

```python
oob_sc = model.oob_score_
print("The Oob Score of the Random Forest Model :",oob_sc)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

prediction = model.predict(X_train)

mse = mean_squared_error(y_train,prediction)
print("The mean squared error for training data: ", mse)

r2_sc = r2_score(y_train,prediction)
print("The r2 error for training data: ", r2_sc)
```

```python
prediction = model.predict(X_test)

mse = mean_squared_error(y_test,prediction)
print("The mean squared error for testing data: ", mse)

r2_sc = r2_score(y_test,prediction)
print("The r2 error for testing data: ", r2_sc)
```

The Model fits perfectly without overfitting with 200 n_estimators as a hyperparameter.

Training this model using a neural network.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# Defining the models
model = Sequential()
model.add(Dense(64, activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='mean_squared_error',metrics=['mean_squared_error'])
```

```python
# Early Stopping
early_stopping=EarlyStopping(
    monitor="val_loss",
    min_delta=0.0001,
    patience=20,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=0,
)
```

```python
history = model.fit(X_train,y_train,validation_split=0.2,epochs=100,batch_size=10,callbacks=[early_stopping])
```
